```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
df = pd.read_csv('signal.csv', parse_dates=['Date'])
```

```python
df_h1 = df[(df['Interval'] == '1h') & (df['Indicator'].str.contains('Support'))]
```

```python
df_h1
```

```python
from binance.client import Client
binance_client = Client()
```

```python
import random
random_index = random.randint(0, df.shape[0])
selected_signal = df.iloc[random_index]
```

```python
selected_signal
```

```python
pair = selected_signal['Pair']
pair
```

```python
start_date = int(selected_signal['Date'].to_pydatetime().timestamp()*1000)
end_date = start_date + (60*60*24*3)*1000 # after 3 days
```

```python
klines_m1 = binance_client.get_historical_klines(pair, Client.KLINE_INTERVAL_1MINUTE, start_date, end_date)
# klines_m1
```

```python
ohlc_data = []
for kline in klines_m1:
    if kline[0]:
        ohlc_data.append([kline[0], float(kline[1]), float(kline[2]), float(kline[3]), float(kline[4]), float(kline[5])])
len(ohlc_data)
```

```python
df_ohlc = pd.DataFrame(ohlc_data, columns=['Date', 'Open', 'High', 'Low', 'Close', 'Volume'])
df_ohlc['Date'] = df_ohlc['Date'].apply(lambda x: pd.to_datetime(x, unit='ms'))
df_ohlc = df_ohlc.set_index('Date')
# df_ohlc
```

```python
# df_ohlc.iloc[-1:].index
```

```python
df_ohlc.dtypes
```

```python
df_ohlc_h1 = df_ohlc.resample('60min').agg({
    'Low': lambda s: s.min(),
    'High': lambda s: s.max(),
    'Open': lambda s: s[0],
    'Close': lambda s: s[-1],
    'Volume': lambda s: s.sum()
})
```

```python
entry_price = df_ohlc.iloc[1]['Open']
```

```python
print("BUY {} at: {}, price {}".format(pair, df_ohlc.iloc[1].name.to_pydatetime().strftime("%m/%d/%Y, %H:%M:%S"), entry_price))
hit = False
for i, row in df_ohlc.iterrows():
    open_price = row['Open']
    if open_price >= entry_price * 1.03:
        print("Hit TP at {}".format(row.name.to_pydatetime().strftime("%m/%d/%Y, %H:%M:%S"))); hit = True
        break;
    if open_price <= entry_price * 0.97:
        print("Hit SL at {}".format(row.name.to_pydatetime().strftime("%m/%d/%Y, %H:%M:%S"))); hit = True
        break;
if not hit: print('Not Hit TP/SL in 3 days')
```
